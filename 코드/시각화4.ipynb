```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
#from pylab import cm

import warnings 
warnings.filterwarnings('ignore')
pd.options.display.float_format = '{:.5f}'.format

from matplotlib import font_manager, rc
font_path = "C:/Windows/Fonts/malgun.ttf"
font = font_manager.FontProperties(fname=font_path).get_name()
rc('font', family=font)

clustering = pd.read_csv("C:/python_project/datasets/gmm_clustering_2.csv")

inst_cluster = clustering[["inst_rep_loanb","cluster"]]
ls_cluster = clustering[["ls_rep_loanb", "cluster"]]

```

```python
plt.figure(figsize=(15,10))
sns.boxplot(x='cluster', y ='ls_rep_loanb', data=ls_cluster , palette='Spectral', linewidth = 3,width = 0.5) 
plt.title('\n   \n',fontsize=23)
#sns.boxplot(x='cluster', y ='inst_rep_loanb', data=inst_cluster) 
plt.xticks(fontsize=15)
```

```python
inst_cluster_total = inst_cluster.groupby('cluster').sum()
ls_cluster_total = ls_cluster.groupby('cluster').sum()
inst_cluster_total
```

```python
clustering_rate = pd.read_csv("C:/python_project/datasets/gmm_clustering_2.csv")

clustering_rate
```

```python
loanb_total = pd.concat([inst_cluster_total, ls_cluster_total], axis=1)
loanb_total['loanb_total'] = loanb_total['inst_rep_loanb'] + loanb_total['ls_rep_loanb']
loanb_total
```

```python
loanb_total['inst_loanb_rate'] = loanb_total['inst_rep_loanb'] / loanb_total['loanb_total'] * 100
loanb_total['ls_loanb_rate'] = 100 - loanb_total['inst_loanb_rate']
loanb_total
```

```python
import numpy as np
fig, ax = plt.subplots(figsize=(30,15))
bar_width = 0.3
index = np.arange(7)

b1 = plt.bar(index - bar_width, loanb_total['inst_loanb_rate'], bar_width, alpha=0.4, color='red', label='inst_loanb_rate')

b2 = plt.bar(index, loanb_total['ls_loanb_rate'], bar_width, alpha=0.4, color='blue', label='ls_loanb_rate')

for rect in b1:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 25)
    
for rect in b2:
    height = rect.get_height()
    plt.text(rect.get_x() + rect.get_width()/2.0, height, '%.1f' % height, ha='center', va='bottom', size = 25)


# x, y    
plt.xlabel('cluster', size = 20)
plt.ylabel('loanb_rate', size =20)
plt.legend()
plt.show()
```

```python
# 

import seaborn as sns

clustering = pd.read_csv("C:/python_project/datasets/gmm_clustering_2.csv")

# inst_cluster = clustering[["inst_rep_loanb","cluster"]]
# ls_cluster = clustering[["ls_rep_loanb", "cluster"]]
sum_inst_clustering = clustering.groupby(['cluster'])['inst_rep_loanb'].sum()

group_names = sum_inst_clustering.index
group_sizes = sum_inst_clustering.values
group_colors = ['#4458a2', '#526fd0', '#6087ff', '#839aff', '#a0adff', '#bac1ff', '#d2d5ff']
group_explodes = (0.0, 0.0, 0.0)
fig = plt.figure(figsize=(10, 10))

plt.pie(group_sizes,  
        labels=group_names, 
        colors=group_colors, 
        autopct='%1.2f%%',
        shadow=True, 
        startangle=90,
        textprops={'fontsize': 14}, counterclock=False)
centre_circle = plt.Circle((0,0),0.50,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.axis('equal')

plt.show()
```
